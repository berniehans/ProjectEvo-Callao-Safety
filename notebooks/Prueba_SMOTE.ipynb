```python
################################################################################
#                                                                              #
#                         PROJECT EVO - VERSION 3.0                            #
#                           ProjectEvo_SMOTE.ipynb                             #
#                                                                              #
################################################################################


# --- 1. SETUP DE ALTO RENDIMIENTO ---
import shutil
import os
print("âš™ï¸ Preparando entorno de ProducciÃ³n...")
if shutil.which("uv") is None:
    !pip install uv -q
!uv pip install --system -q torch transformers pandas numpy scikit-learn tqdm imbalanced-learn matplotlib seaborn

import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gc
from torch.utils.data import DataLoader, TensorDataset
from transformers import BertTokenizer, BertModel, logging
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, f1_score, confusion_matrix
from imblearn.over_sampling import SMOTE
from google.colab import drive
from tqdm.notebook import tqdm

# ConfiguraciÃ³n
logging.set_verbosity_error()
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"ðŸš€ ACELERADOR: {torch.cuda.get_device_name(0)}")

# --- 2. CARGA MASIVA DE DATOS (TOTALIDAD) ---
RUTE_ARCHIVO = 'https://raw.githubusercontent.com/berniehans/DataCallao/main/v2_dataset_callao.csv'

print("\nðŸ“– Leyendo base de datos completa...")
# Leemos TODO sin recortar
df_full = pd.read_csv(RUTE_ARCHIVO, sep=';', encoding='latin1', usecols=['DESCRIPCION_OCURRENCIA', 'TIPOLOGIA'])
df_full = df_full.dropna().reset_index(drop=True)

print(f"âœ… Total Registros en bruto: {len(df_full)}")

# CodificaciÃ³n
encoder = LabelEncoder()
y_full = encoder.fit_transform(df_full['TIPOLOGIA'])
CLASES_NAMES = encoder.classes_
NUM_CLASSES = len(CLASES_NAMES)
X_text_full = df_full['DESCRIPCION_OCURRENCIA'].astype(str).tolist()

print(f"ðŸŽ¯ Clases ({NUM_CLASSES}): {CLASES_NAMES}")

# --- 3. BERT EMBEDDINGS (BATCH PROCESSING PARA 100K) ---
# Al ser muchos datos, procesamos por lotes para no saturar la GPU
print("\nðŸ¤– Generando vectores para la TOTALIDAD de los datos...")
tokenizer = BertTokenizer.from_pretrained('dccuchile/bert-base-spanish-wwm-cased')
bert_model = BertModel.from_pretrained('dccuchile/bert-base-spanish-wwm-cased').to(DEVICE)
bert_model.eval()

embeddings_list = []
BATCH_SIZE = 128 # Batch mÃ¡s grande para ir rÃ¡pido

for i in tqdm(range(0, len(X_text_full), BATCH_SIZE), desc="Vectorizando"):
    batch_text = X_text_full[i : i + BATCH_SIZE]
    inputs = tokenizer(batch_text, return_tensors="pt", padding=True, truncation=True, max_length=64).to(DEVICE)
    with torch.no_grad():
        out = bert_model(**inputs).pooler_output.cpu().numpy()
        embeddings_list.append(out)

X_full_emb = np.concatenate(embeddings_list, axis=0)

# Limpieza crÃ­tica de memoria
del bert_model, tokenizer, inputs, embeddings_list
torch.cuda.empty_cache()
gc.collect()

# --- 4. DIVISIÃ“N ESTRATÃ‰GICA (TRAIN vs INFERENCE POOL) ---
print("\nâœ‚ï¸ Separando datos de prueba (Hold-Out)...")
# Separamos el 20% de los datos como "Inferencia/Test".
# El modelo NO VERÃ estos datos durante el entrenamiento ni el SMOTE.
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(
    X_full_emb, y_full, test_size=0.20, random_state=42, stratify=y_full
)

print(f"   ðŸ“‚ Set de Entrenamiento: {len(X_train_raw)} registros")
print(f"   ðŸ”’ Set de Inferencia (Test): {len(X_test_raw)} registros (Nunca vistos)")

# --- 5. SMOTE SOLO EN ENTRENAMIENTO ---
# Es vital aplicar SMOTE solo al train para no falsear los resultados del test
print("\nâš–ï¸ Aplicando SMOTE al set de Entrenamiento...")
# Estrategia: No inflar la clase mayoritaria al infinito.
# Vamos a balancear las minoritarias para que igualen a la media o un tope razonable.
# SMOTE por defecto iguala a la mayoritaria. Si es muy grande (>50k), podrÃ­a ser lento.
# En tu caso 100k total -> mayoritaria ~30k. Es manejable.
smote = SMOTE(random_state=42, k_neighbors=3)
X_train_bal, y_train_bal = smote.fit_resample(X_train_raw, y_train_raw)

print(f"   TRAIN original: {len(X_train_raw)} -> TRAIN balanceado: {len(X_train_bal)}")

# NormalizaciÃ³n (Fit en train, Transform en test)
scaler = StandardScaler()
X_train_bal = scaler.fit_transform(X_train_bal)
X_test_scaled = scaler.transform(X_test_raw) # Importante: usar el scaler del train

# A GPU
X_train_gpu = torch.tensor(X_train_bal, dtype=torch.float32).to(DEVICE)
y_train_gpu = torch.tensor(y_train_bal, dtype=torch.long).to(DEVICE)
X_test_gpu = torch.tensor(X_test_scaled, dtype=torch.float32).to(DEVICE)
# y_test queda en CPU para mÃ©tricas

# --- 6. ARQUITECTURA GANADORA (RESULTADO DEL GA) ---
BEST_NEURONS = 448
BEST_DROPOUT = 0.16
BEST_LR = 0.00273

print(f"\nðŸ§  Construyendo Cerebro Optimizado ({BEST_NEURONS} neuronas, Drop {BEST_DROPOUT})...")

model = nn.Sequential(
    nn.Linear(768, BEST_NEURONS),
    nn.BatchNorm1d(BEST_NEURONS),
    nn.ReLU(),
    nn.Dropout(BEST_DROPOUT),
    nn.Linear(BEST_NEURONS, NUM_CLASSES)
).to(DEVICE)

optimizer = torch.optim.Adam(model.parameters(), lr=BEST_LR)
criterion = nn.CrossEntropyLoss()

# --- 7. ENTRENAMIENTO FINAL A ESCALA ---
print("ðŸ”¥ Entrenando con la totalidad de datos de entrenamiento...")
EPOCHS = 35
BATCH_SIZE_TRAIN = 256 # Batch mÃ¡s grande para velocidad

loss_hist = []
model.train()

for epoch in tqdm(range(EPOCHS), desc="Training"):
    perm = torch.randperm(len(X_train_gpu))
    epoch_loss = 0

    for i in range(0, len(X_train_gpu), BATCH_SIZE_TRAIN):
        idx = perm[i : i + BATCH_SIZE_TRAIN]
        optimizer.zero_grad()
        out = model(X_train_gpu[idx])
        loss = criterion(out, y_train_gpu[idx])
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    loss_hist.append(epoch_loss / (len(X_train_gpu)/BATCH_SIZE_TRAIN))

# --- 8. PRUEBA DE INFERENCIA (EL EXAMEN FINAL) ---
print("\nðŸ•µï¸â€â™‚ï¸ Ejecutando Inferencia sobre datos NO VISTOS...")
model.eval()

with torch.no_grad():
    # El modelo predice sobre los 20k datos "vÃ­rgenes"
    outputs = model(X_test_gpu)
    y_pred = torch.argmax(outputs, dim=1).cpu().numpy()

# --- 9. RESULTADOS Y VISUALIZACIÃ“N ---
print("\n" + "="*60)
print(f"ðŸ† RESULTADOS FINALES: VALIDACIÃ“N CON {len(y_test_raw)} REGISTROS")
print("="*60)

# MÃ©tricas Globales
print(classification_report(y_test_raw, y_pred, target_names=CLASES_NAMES))

# Matriz de ConfusiÃ³n
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(loss_hist, label="Error de Entrenamiento")
plt.title("Estabilidad del Aprendizaje")
plt.xlabel("Epochs")
plt.grid(True, alpha=0.3)
plt.legend()

plt.subplot(1, 2, 2)
cm = confusion_matrix(y_test_raw, y_pred, normalize='true')
sns.heatmap(cm, annot=True, fmt=".2f", cmap="Blues",
            xticklabels=CLASES_NAMES, yticklabels=CLASES_NAMES)
plt.title("PrecisiÃ³n Real por Clase (Test Set)")
plt.tight_layout()
plt.show()

# Guardar modelo (Opcional, para tu tesis)
print("ðŸ’¾ Guardando modelo entrenado como 'modelo_tesis_v3.pth'...")
torch.save(model.state_dict(), "modelo_tesis_v3.pth")
```
